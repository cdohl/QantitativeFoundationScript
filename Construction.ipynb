## 1.Numbers
Cells come in a variety of shape and sizes. We start with the bacterium Escherichia coli (E. coli) as our standard ruler. Very simple estimates for the genome size, numbers of lipids, proteins, and ribosom content. Here we discuss scales (sizes, areas, and volumes) using powers of 10.


Figure 1.1 <img src="figs/Ecoli.png" style="width:300px" /> (taken from Physical Biology of the Cell page 37)

Figure 1.1 depicts on top an image of an E. coli taken with an atomic force microscope, revealing the main body covered with flagella and pili. Below is an electron microscope picture revealing DNA content of the cell. The length of the cell is about $2\,\mu$m$=2\cdot 10^{-6}\,$m and the diameter between $0.5\,\mu$m$=5\cdot 10^{-7}\,$m and $1\,\mu$m. We can estimate the volume of the cell abstracting it as a [cylinder](https://en.wikipedia.org/wiki/Cylinder) with radius of $r=0.5\,\mu$m and length $l=2\,\mu$m:

$$
V=\pi r^2 l \approx 3\cdot (0.5)^2\cdot 2\cdot (\mu\,\mathrm{m})^3=1.5\,(\mu\mathrm{m})^3 \approx 1 \,(\mu\mathrm{m})^3
$$

and for the surface area (ignoring the caps) we obtain

$$
A=2\pi r \cdot l\approx 6 \cdot 0.5 \cdot 2 \cdot (\mu\,\mathrm{m})^2=6\, (\mu\mathrm{m})^2
$$

Now let's look at the content of the cell. The E. coli contains proteins, ribosomes, lipids (on the cell membran), water, inorganic ions. The volume of the bacteria is $1 \,\mu\mathrm{m}^3$, which is $1\,$fl (femtoliter) (check!) or if we assume the density of water $1\,$pg (picogram).  

Experiments have revealed the dry mass of E. coli is one third of the mass weight, and one half of that are proteins, i.e.  $1\,$pg$ / 6 \approx 0.15\,$pg or $1.5 \cdot 10^{-13}\,$g or $1.5 \cdot 10^{-15}\,$kg. The number of molecules in that mass can be obtained from an estimation of the mass of one molecules. Assume that each protein consists *on average* of 300 [amino acids](https://en.wikipedia.org/wiki/Amino_acid), each having a mass of 100 Da, [Dalton](https://en.wikipedia.org/wiki/Dalton_unit).
The mass of one atomic unit (written as **u**) or Dalton (written as **Da**) (that is the more common phrase used in Biology, but both are the same) is approx. $1.6\cdot 10^{-27}\,$kg or $1.6\cdot 10^{-24}\,$g. It is important to remember the numeric value of one Da, as it will haunt you for the rest of your studies.
Looking at the list of amino acids and their molecular weight, e.g. on [Wikipedia](https://en.wikipedia.org/wiki/Amino_acid#Physicochemical_properties) we can estimathe the weight of one as approx. 100 Da, thus the average weight of one amino acid is about 30,000 Da or $3\cdot 10^{4}\,\mathrm{Da} \cdot 1.6\cdot 10^{-24}\,\mathrm{g}\approx 5 \cdot 10^{-20}\,$g. Therefore the number of proteins is

$$N_\mathrm{protein}=\frac{\text{total protein mass}}{\text{mass per protein}}=\frac{15\cdot 10^{-14}\,\mathrm{g}}{5\cdot 10^{-20}\,\mathrm{g}}\approx 3\cdot 10^6\,\mathrm{proteins}
$$


Proteins are synthesized in cellular machines termed ribosomes. It is known that about 20\% of the proteins in a cell are ribosomal proteins and the average mass of a ribosom is 2.5 MDa. 30\% of that are proteins. We can thus determine the number of ribosoms as



$$N_\mathrm{ribosome}=\frac{0.2 \cdot 15\cdot 10^{-14}\,\mathrm{g}}{0.3\cdot 2.5\cdot 10^6 \mathrm{Da}}=\frac{3\cdot 10^{-14}\,\mathrm{g}}{0.75\cdot 10^6\cdot 1.6\cdot 10^{-24}\,\mathrm{g}}\approx 25,000\,\mathrm{ribosomes}\quad .
$$

We can also estimate the number of water molecules in a cell knowing that the dry mass of an E. coli is 30\% of the wet mass, i.e. 70\% of the mass of a cell is water. The water molecule consists of two hydrogen atoms and one oxygen atom, thus its mass is 18Da and the number of molecules is

$$N_\mathrm{H_2O}=\frac{0.7\cdot 1\,\mathrm{pg}}{18\cdot 1.6\cdot 10^{-24}\,\mathrm{g}}\approx 2.4 \cdot 10^{10}\,\text{water molecules}\quad .
$$

## 2. Crowded interior

Figure 1.2 <img src="figs/Ecoli_inner.png" style="width:600px" /> (taken from Physical Biology of the Cell page 42)

Figure 1.2 depicts a small part of the E. coli cell with an approximate number of the different molecules. When respecting the cellular census the cell is a crowded with molecules. This needs to be taken into account when looking at transport processes (diffusion) within the cell.

We can estimate the concentration of molecules in the cell. Therefore we need to introduce a unit for measuring elementary entities such as atoms, molecules, ions, and other kinds of particles. We define this unit as one mole and it contains exactly $6.02214076\cdot 10^{23}$ entities. This number was chosen such that one mole is the number of atoms in 12 grams of Carbon ($^{12}$C).

The concentration of entities in a solution is commonly expressed by its molar concentration, defined as the amount of entities per unit volume of solution. Commonly the concentration of one  mole per litre (mol/L) is used and is abbreviated with the letter M. The volume of the E. coli is $1\,\mu$m$^3=10^{-18}\,$m$^3=1\,$fL. Thus the concentration of 1 molecule in the E. coli is $10^{15}$ molecules per liter.

$$
\frac{10^{15} \,\text{Molecules per L}}{6.022\cdot 10^{23} \,\frac{\text{Molecules per L}}{\text{M}}}\approx 1.7\cdot 10^{-9}\,\text{M}\approx 2\,\text{nM}
$$

As a rule of thumb, a concentration of 2nM implies one molecule within the volume of an E. coli bacteria.
Consider that the molecules are organized on the center of cubes arranged as lattice in 3 dimensional space. The sidelength of each cube is $d$. Then the distance between two neighbouring molecules also $d$ and the concentration is 

$$c=\frac{\text{Number of molecules}}{\text{Volume}}=\frac{1}{d^3}$$

or the mean distance between molecules is

$$d=\frac{1}{\sqrt[3]{c}}=c^{-1/3}\quad .$$



```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 18})

c=np.logspace(-9,-6,100) #vector from 1nM to 1000nM
n=c/1e-9/1.7 #1nM contains 1/1.7 cells
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(12,4),constrained_layout=True)
ax1.loglog(c*1e9,n)
ax1.set_xlabel('concentration (nM)')
ax1.set_ylabel('no. of molecules')
c=np.logspace(-9,-3,100)
d=c**(-1./3.)
ax2.loglog(c*1e9,d)
ax2.set_xlabel('concentration (nM)')
ax2.set_ylabel('d (nm)');
```

## 1.3 Timing of E. coli

Here we look at the timescales of processes occuring in cells. The E. coli bacteria to divide and split needs to double its length and therefore double its volume. Under optimal growth condition (nutrition and temperature) this happens within 40-50 minutes. For a rough estimate let it fix to $50\cdot 60\,$s$=3000\,$s. The number of carbon atoms in an E. coli cell is about $10^{10}$. Thus for cell division to occur, $10^{10}$ C atoms need to be taken up over 3000 s. These is occuring through the uptake of glucose with 6 carbon atoms, thus the rate of glucose uptake is $10^{10}/(6\cdot 3000)\,$Glucose molecules/s which is roughly $5\cdot 10^{5}\,$ glucose molecules/s. The carbon molecules are important for genome, protein synthesis, and lipids. Besides the cell consists of water. We have estimated above $2\,10^{10}$ water molecules per cell and can then estimate the rate of water uptake as

$$ \frac{\mathrm{N_{\text{H}_2\text{O}}}}{\tau_\text{cell}}\approx \frac{2\cdot 10^{10}\, \text{water molecules}}{3000\,\mathrm{s}}\approx 7 \cdot 10^6 \,\text{water molecules/s}\quad .
$$
