# Rotational motion, energy and angular momentum

## 6.1 Rotational dynamics
 
 Let us describe the rotatation of a disk with radius $R$ rotating around a center point C.
 
 The angle of of rotation is called $\Theta$, and the tangential velocity of the rim of the disk is $v=\omega R=\dot{\theta}R$.
 
<img src="figs/8n.1_1.png" width=240px>

 If we allow for an acclerated motion, i.e. the angular velocity is allowed to change, then we have a 
 $a_\mbox{tan}=\dot{\omega}R=\ddot{\Theta}R=\alpha R$.
 
 Here $\alpha$ is the angular acceleration.
 
 We can connect the linear and the circular motion by replacing the variables as shown in the table below.
 
| linear motion | circular motion |
| --- | ---|
| x |  $\Theta$ | 
| v |  $\omega$ | 
| a |  $\alpha$ | 
| m |  I | 


Thus the position $x$ in linear motion becomes the angle $\Theta$. We can rewrite the formulas for motion with constant acceleration first as linear motion:

$$x(t)=x_0+v_0 t + \frac{1}{2}a t^2$$

and for circular motion:

$$\Theta(t)=\Theta_0+\omega t + \frac{1}{2}\alpha t^2\quad .$$

And similiarily for the velocity in linear motion:

$$v(t)=v_0 + a t$$

compared to 

$$\omega(t)=\omega_0+\alpha t\quad .$$


## 6.2 Rotational energy and moment of inertia

Let us now describe the kinetic energy of a rotating disk. The disk rotates through a stationary axis C.


<img src="figs/8n.1_2.png" width=240px>


We start first with the kinetic energy of the mass element $m_i$ due to the rotation of the disk. This is straightforward:

$$K_i=\frac{1}{2}m_i v_i^2$$

Here the velocity $v_i$ is a function of the location on the disk. This can be simplified knowing that $v=\omega r_i$ to

$$K_i=\frac{1}{2}m_i \omega^2 r_i^2$$

The kinetic energy of the whole disk is then

$$K_\mbox{disk}=\frac{\omega^2}{2}\sum m_i r_i^2$$

The sum is named the moment of inertia of the rotating disk, i.e. 

$$I_C=\sum m_i r_i^2\quad .$$

Here $I_c$ stands for the moment of inertia of an object rotating through the axis C. This is important to remember, that the moment of inertia depends on the specific axis the object is rotating through.

We can  writt $I_C$ as an integral $I_C=\int_V r^2 dV$ and solved for. This is a mathematical excercise (which will do below) yet for now let us use the result for the disk that is

$$I_{C,\mbox{disk}}=\frac{1}{2}m R^2$$


Some other momemnt of inertias are

for a solid sphere where the rotation axis goes throuhgh the center of the object

<img src="figs/8n.1_3.png" width=240px>

$$I_{sphere}=\frac{2}{5} m R^2$$

and for a thin rod of length $l$ and an axis goingt through the center of the object:

<img src="figs/8n.1_4.png" width=240px>

$$I_{rod}=\frac{1}{12} m l^2$$

## 6.3 Parallel axis theorem

<img src="figs/8n.1_5.png" width=250px>


The moment of inertia of an object rotating along axis $l$ going through the center of mass be $I_l$. If the object is rotated through an axis parallel to $l$ at a distance $d$, thus through $l'$ as shown in the figure above the moment of inertia can be obtained from the parallel axis theorem:

$$I_l'=I_l+m d^2\quad ,$$

where $m$ is the mass of the object.




## 8.4. Rotational and potential energy

Let us first have a look at the rolling motion

### Rolling motion
A rolling motion consists of the translation of the center of mass and a rotation around an axis through the center of gravity of the rolling body. In ideal rolling without sliding, both speeds are given by the relationship $v=\omega\,R$.


<img src="figs/8n.1_6.png" width=350px>


### Calculation of the moment of inertia for a cylinder (example)


A cylinder with constant density that rotates around the axis of symmetry $\vec{\omega}=\omega\,\vec{e}_z$ through the center $C: x=y=z=0$:

<img src="figs/8.3_2.png" width=200px>

$$I_{C,z}=\int r^2 \,\mathrm{d}m\quad .$$

At constant density, $\mathrm{d}m=\rho\,\mathrm{d}V$ and

$$I_{C,z}=\rho\int r^2 \,\mathrm{d}V\quad .$$

A suitable volume element must be used for the integration over $\mathrm{d}V$. Cylindrical coordinates are suitable here. These are easy to understand:

<img src="figs/8.3_3.png" width=250px>

The surface element $\mathrm{d}A$ of a cross-section of the cylinder is

$$\mathrm{d}A=\underbrace{r\,\mathrm{d}\varphi}_{\mathrm{arc length}}\,\mathrm{d}r$$

The volume then only takes into account the depth in the $z$-direction:

$$\mathrm{d}V=r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

$$I_{C,z}=\rho\int r^2 \,\mathrm{d}V\quad=
\rho\int_{-h/2}^{h/2}\int_0^R\int_0^{2\pi}r^2\,r\,\mathrm{d}\varphi\,\mathrm{d}r\,\mathrm{d}z$$

Integration over $\mathrm{d}\varphi$ results in $2\pi$

$$I_{C,z}=2\pi\,\rho\int_{-h/2}^{h/2}\int_0^R r^3\mathrm{d}r\,\mathrm{d}z$$

Integration over $\mathrm{d}z$ results in $h$

$$I_{C,z}=2\pi\,\rho h \int_0^R r^3\mathrm{d}r$$

And finally via $r$

$$I_{C,z}=2\pi\,\rho h \left[\frac{r^4}{4}\right]_0^R=\underbrace{\rho \pi R^2 h}_{\rho\,V=m} \,\frac{1}{2}R^2=\frac{1}{2}m R^2$$



### Speed of a rolling cylinder
We can determine the speed of a cylinder rolling down an inclined plane using the law of conservation of energy. To do this, we assume that no frictional work is performed (the contact point between the cylinder and the plane is at rest).

<img src="figs/8.3_5.png" width=350px>


If the cylinder is at rest at the beginning, then 

$$E_{pot}=m g h$$

Then it speeds up. We have two kinetic energies due to the translation and the rotation:

$$E_{kin}=\frac{1}{2} m v^2+\frac{1}{2}I_C\,\omega^2$$

After the cylinder has rolled down the height $h$, we have converted the potential energy into kinetic energy, so the velocity is

$$v=\sqrt{\frac{2 g h}{1+\frac{I_C}{m R^2}}}$$

In the case of a cylinder with $I_C=\frac{1}{2}m R^2$

we obtain a velocity of 

$$v=\sqrt{\frac{4}{3}g\,h}<\sqrt{2 g h}$$

The velocity is less than the velocity of a mass point falling freely a height $h$, namely $v=\sqrt{2 g h}$. Why is this?
