# 4. Work and mechanical energy

## 4.1 Work

A force $\vec{F}$ that moves a body a certain distance $\mathrm{d}\vec{r}$ performs work:

$$\mathrm{d}W=\vec{F}\cdot\mathrm{d}\vec{r}$$

The total work involved in a displacement

$$\int_0^{W_{12}}=\int_{\vec{r}_1}^{\vec{r}_2}\vec{F}\cdot\mathrm{d}\vec{r}$$

$$W_{12}=\int_{\vec{r}_1}^{\vec{r}_2}\vec{F}\cdot\mathrm{d}\vec{r}\tag{4.1.1}$$

The symbol for work is $W$, the unit is $\left[W\right]=\mathrm{J}$ which relates to Joule.

Remeber that we are dealing here with a scalar product:

$$ \vec{F}\cdot\mathrm{d}\vec{r}=|\vec{F}|\,|\mathrm{d}\vec{r}|\,\cos (\vec{F},\mathrm{d}\vec{r})$$

The work is maximum: $\vec{F}\parallel\mathrm{d}\vec{r}$

The work is minimum: $\vec{F}\perp\mathrm{d}\vec{r}$


<b>Displacement work (against frictional force)</b>

<img src="figs/6.1_1.png" width=300px>

$$\vec{v}=\mathrm{const.}=v_x\,\vec{e}_x$$

$$\mathrm{d}\vec{r}\parallel\vec{e}_x$$

$$\vec{F}_{dis}=F_x\,\vec{e}_x+F_y\,\vec{e}_y$$

\begin{eqnarray}
F_x&=&\left|\vec{F}_{dis}\right|\,\cos\beta\\
F_y&=&\left|\vec{F}_{dis}\right|\,\sin\beta
\end{eqnarray}

If $\vec{v}=\mathrm{const.}$ then

$$\vec{F_x}\,\vec{e}_x+\vec{F}_{GL}$$
$$\vec{F_x}\,\vec{e}_x=-\vec{F}_{GL}$$

and the path of the object is

$$\mathrm{d}\vec{r}=\vec{e}_x\,\mathrm{d}x$$

Then we can write the integral Eq. (4.1.1) as: 


\begin{eqnarray}
W_{12}&=&\int_{x_1}^{x_2}\vec{F}_{dis}\cdot\mathrm{d}\vec{r}\\
&=&\int_{x_1}^{x_2}\left(F_x\,\vec{e}_x+F_y\,\vec{e}_y \right)\vec{e}_x\,\mathrm{d}x\\
&=&\int_{x_1}^{x_2} F_x\,\vec{e}_x\cdot\vec{e}_x\,\mathrm{d}x + \int_{x_1}^{x_2} F_y\,\underbrace{\vec{e}_y\cdot\vec{e}_x}_{\vec{e}_x\cdot\vec{e}_y=0}\,\mathrm{d}x
\end{eqnarray}

Because $\vec{e}_x \perp \vec{e}_y$ is

$$W_{12}=\int_{x_1}^{x_2}F_{x}\,\mathrm{d}x$$

and with $F_x=\mathrm{const.}$ we obtain

$$W_{12}=F_x\,\left(x_2-x_1\right)\quad .$$

Only the components $\vec{e}_x$ of $\vec{F}_{dis}$ perform work.

<b>Work against gravity (close to Earth)</b>


*Close to Earth* means that the weight (force) is constant. This is generally not the case. We will get to know a generalized law for mass attraction later.


<img src="figs/6.1_2.png" width=300px>

The force necessary to lift a mass

\begin{eqnarray}
\vec{F}_{lift}&=&-\vec{F}_g\\
&=&m \, g\, \vec{e}_z
\end{eqnarray}

This happens in absence of acceleration, i.e *infinitely slow*.

$$W_{12}=\int_0^h \vec{F}_{lift}\cdot\mathrm{d}\vec{r}=
\int_0^h m\,g\,\underbrace{\vec{e}_z\cdot\vec{e}_z}_{=1}\,\mathrm{d}z$$

and we obtain

$$W_{12}=m\,g\,h\quad .$$

This holds only for a constant mass $m$ and constant $g$.


<b>Acceleration work</b>

accelerating force: $\vec{F}_B=m\,\vec{a}$
  $$\vec{F}_B=m\,\frac{\mathrm{d}\vec{v}}{\mathrm{d}t}$$
  
$$W_B=W_{12}=\int_{\vec{r}_1}^{\vec{r}_2} m\,\frac{\mathrm{d}\vec{v}}{\mathrm{d}t} \cdot \mathrm{d} \vec{r}=\int_{\vec{r}_1}^{\vec{r}_2} m\,\frac{\mathrm{d}\vec{r}}{\mathrm{d}t} \cdot \mathrm{d} \vec{v}$$

because of

$$\frac{\mathrm{d}\vec{r}}{\mathrm{d}t}=\vec{v}$$

we write the integral

\begin{eqnarray}
W_{12}&=&\int_{\vec{v}_1(\vec{r}_1)}^{\vec{v}_2(\vec{r}_2)}m\,\vec{v}\cdot\mathrm{d}\vec{v}\\
&=&\left.\frac{m}{2}\vec{v}^2\right|_{\vec{v}_1}^{\vec{v}_2}=
\frac{m}{2}\left({\vec{v}_2}^2-{\vec{v}_1}^2\right)
\end{eqnarray}

This means that the acceleration work for a body from speed $\vec{v}_1$ to speed $\vec{v}_2$ is 

$$W_B=\frac{m}{2}\left({\vec{v}_2}^2-{\vec{v}_1}^2\right)$$



## 4.2 Power

### Instantaneous power

$$P=\frac{\mathrm{d}W}{\mathrm{d}t}$$

Unit: $\left[P\right]=1\,$W 


\begin{eqnarray}
d\mathrm{W}&=&P\,\mathrm{d} t\\
W_{12}&=&\int_{t_1}^{t_2}\,P\,\mathrm{d}t
\end{eqnarray}

<img src="figs/6.2_1.png" width=400px>

The average power $\bar{P}$ in the time interval $\Delta t$ is obtained by

$$\bar{P}=\frac{W_{12}}{\Delta t}=\left< P\right>$$


## 4.3 Mechanical energies

### 4.3.1 Kinetic energy

* State of bodies $\rightarrow$ Energy is assigned
* depends on the velocity
* $E_{kin}=\frac{m}{2} v^2$
* Kinetic energy can be changed by acceleration work

  \begin{eqnarray}
  E_{kin}^{(2)}&=&\frac{m}{2} v_2^2\\
  E_{kin}^{(1)}&=&\frac{m}{2} v_1^2\\
  \Delta E_{kin}=E_{kin}^{(2)}-E_{kin}^{(1)}&=&\frac{m}{2}\left(v_2^2-v_1^2\right)
  \end{eqnarray}
  
 The difference in the kinetic energy $\Delta E_{kin}=W_{12}$ is the acceleration work.
 
### 4.3.2 Potential energy

* Is stored energy that can be released again
* Definition of potential energy
  $$E_{pot}=-\int_{\vec{r}_0}^{\vec{r}} \vec{F_{kon}}\,\mathrm{d}\vec{r}$$
  $\vec{r}_0$ is the zero point and can, for example, be filled with the potential energy of zero, i.e. $E_{pot}(\vec{r}=\vec{r}_0)=0$. The force in the integral must be conservative in order to be able to assign a potential energy to it, e.g. spring energy:
  
  
  
* $E_{pot}$ for spring: $E_{pot}(x=0)=0$

  \begin{eqnarray}
  E_{pot}&=&-\int_0^{x_2}\vec{F}_F\cdot\mathrm{d}\vec{r}\\
  &=&-\int_0^{x_2}-k\,x\,\vec{e}_x\cdot\vec{e}_x\,dx\\
  E_{pot}&=&\frac{k}{2}\,x_2^2
  \end{eqnarray}
  
* $E_{pot}$ for lifting a body to height $h$

  \begin{eqnarray}
  E_{pot}&=&-\int_0^{h}\vec{F}_G\cdot\mathrm{d}\vec{r}\\
  &=&-\int_0^{h}-m\,g\,\vec{e}_z\cdot\vec{e}_z\,dz\\
  E_{pot}&=&m\,g\,h
  \end{eqnarray}

### 4.3.3 Law of conservation of energy in mechanics

Mechanical energy ($E_{kin}$ and $E_{pot}$) can be converted into each other.

⇒ the increase in $E_{kin}$ comes from the supply of $E_{pot}$

⇒ the change over time are opposite to each other

$$\frac{\mathrm{d}E_{kin}}{\mathrm{d}t}=-\frac{\mathrm{d}E_{pot}}{\mathrm{d}t}$$

$$\frac{\mathrm{d}\left(E_{kin}+E_{pot}\right)}{\mathrm{d}t}=0$$

Or to put it another way: 

The conservation of energy applies to closed systems:

$$E_{kin}+E_{pot}=\mathrm{const.}=E_{ges}$$
