## 5 Dynamics of point mass systems

### 5.1 Law of conservation of momentum

The momentum of a body is 

$$\vec{p}=m\,\vec{v}$$

and the unit is $\left[p\right]$=kg m s$^{-1}$.


A force must act for a change in momentum:

$$\vec{F}=m\,\vec{a}=m\,\frac{\mathrm{d}\vec{v}}{\mathrm{d}t}=\frac{\mathrm{d}\left(m\,\vec{v}\right)}{\mathrm{d}t}=
\frac{\mathrm{d}\vec{p}}{\mathrm{d}t}$$

This means that if no force acts on a body, the momentum remains constant. We can extend the concept of momentum to system of point masses. The total momentum results from the vector addition of the individual momenta $\vec{p}_i$, i.e.

$$\vec{p}_{ges}=\sum_i \vec{p}_i=\sum_i m_i\vec{v}_i$$

There applies:

<b> In a closed system in which only internal forces act, the total momentum $p_{ges}$ remains constant.</b>

Internal forces are the forces between the bodies of the system. External forces act on the system from the outside. 


Example:

We consider the movement of a closed system  on a frictionless support. Initially, two masses $m_1$ and $m_2$ are connected with a spring under a spring force via a cable. The masses are at rest, so the total momentum is $p_{ges}=0$.

<img src="figs/7.1_1.png" width=450px>

Then the rope is cut and the masses move apart. No external forces act, the total momentum remains the same.

$$p_{ges}=m_1\,\vec{v_1}+m_2\,\vec{v_2}=0\tag{5.1}$$

The movement is one-dimensional $\vec{v_1}=v_1\vec{e}_x$ and $\vec{v_2}=v_2\vec{e}_x$, therefore applies:

$$\left|\frac{v_1}{v_2}\right|=\frac{m_2}{m_1}$$

If both masses are equal, $m_1$ and $m_2$ move away from each other with the same amount of velocity.

We can further transform Eq. (5.1):

$$m_1\,v_1+m_2\,v_2=p_{ges}\tag{5.2}$$

$p_{ges}$ was zero in the above example, but in general for a system it can assume any constant value.

Inserting the definition of velocity in Eq. (5.2) results in

\begin{eqnarray}
m_1\frac{\mathrm{d}x_1}{\mathrm{d}t}+m_2\frac{\mathrm{d}x_2}{\mathrm{d}t}&=&p_{ges}\\
\frac{\mathrm{d}}{\mathrm{d}t}\left(m_1\,x_1+m_2\,x_2\right)&=&p_{ges}\tag{5.3}
\end{eqnarray}

Now we divide both sides of Eq. (5.3) by the sum of the mass $m_1+m_2$

\begin{eqnarray}
\frac{1}{m_1+m_2}\frac{\mathrm{d}}{\mathrm{d}t}\left(m_1\,x_1+m_2\,x_2\right)&=&\frac{p_{ges}}{m_1+m_2}\\
\frac{\mathrm{d}}{\mathrm{d}t}\underbrace{\left(\frac{m_1\,x_1+m_2\,x_2}{m_1+m_2}\right)}_{x_s}&=&\mathrm{konst.}\tag{5.4}
\end{eqnarray}

#### Center of mass

The expression in the brackets in Eq. (5.4) is constant in time and is therefore called a *conservation quantity*. $x_s$ is a coordinate (with unit length) and points to the center of mass of the closed system. In general, it follows from the conservation of momentum that <b>the center of mass $\vec{r}_s$ of the closed system remains constant</b>.

$$\vec{r}_s=\frac{\sum_i m_i\,\vec{r}_i}{\sum_i m_i}=\mathrm{konst.}$$


The center of mass can be determined for infinitesimally small elements $\mathrm{d}m$ via an integral:


#### Movement of the center of mass

\begin{eqnarray}
\vec{v}_s&=&\dot{\vec{r}_s}=\frac{\mathrm{d}\vec{r}_s}{\mathrm{d}t}\\
&=&\frac{1}{m_{ges}}\sum_{i=1}^N m_i\,\dot{\vec{r}_i}=\frac{1}{m_{ges}}\underbrace{\sum_{i=1}^N m_i\,\vec{v}_i}_{\vec{p}_{ges}}
\end{eqnarray}

Accordingly, the velocity of the center of mass is

$$\vec{v}_s=\frac{1}{m_{ges}}\vec{p}_{ges}$$

<b>The center of gravity of a system moves as if the entire mass were united in the center of gravity</b>.


### 5.2 Laws of impact

* Momentum before and after the impact is a conserved quantity

$$m_1\,\vec{v}_1+m_2\,\vec{v}_2=m_1\,{\vec{v}_1}'+m_2\,{\vec{v}_2}'$$

* Law of conservation of energy

$$ \frac{1}{2}m_1\,{\vec{v}_1}^2+\frac{1}{2}m_2\,{\vec{v}_2}^2=
\frac{1}{2}m_1\,{{\vec{v}_1}'}^2+\frac{1}{2}m_2\,{{\vec{v}_2}'}^2+Q$$

$Q$ is energy other than $W_{kin}$ and $W_{pot}$.


### 5.2.1 Elastic collisions


$Q=0$


<img src="figs/7.2_1.png" width=400px>

Straight central elastic collisions if the direction of movement of $m_1$ and $m_2$ is in the line connecting the centers of mass.

$$\frac{1}{2}m_1\,{v_1}^2+\frac{1}{2}m_2\,{v_2}^2=
\frac{1}{2}m_1\,{{v_1}'}^2+\frac{1}{2}m_2\,{{v_2}'}^2$$

$$\frac{m_1}{2}\left({v_1}^2-{v_1'}^2\right)=\frac{m_2}{2}\left({v_2'}^2-{v_2}^2\right)\tag{5.2.1}$$

$$m_1\,v_1+m_2\,v_2=m_1\,{v_1}'+m_2\,{v_2}'$$

$$m_1\left(v_1-v_1'\right)=m_2\left(v_2-v_2'\right)\tag{5.2.2}$$

Dividing Eq. (5.2.1) by (5.2.2) results in

$$v_1+{v_1}'=v_2+{v_2}'\quad .$$

The two equations can be used to determine the unknown ${v_1}'$ and ${v_2}'$.

$${v_1}'=\frac{\left(m_1-m_2\right)\,v_1+2m_2\,v_2}{m_1+m_2}$$

$${v_2}'=\frac{\left(m_2-m_1\right)\,v_2+2m_1\,v_1}{m_1+m_2}$$

### 5.2.2 Non-elastic collisions

$$Q>0$$

Conservation of energy:

$$\frac{1}{2}m_1\,{v_1}^2+\frac{1}{2}m_2\,{v_2}^2=
\frac{1}{2}m_1\,{{v_1}'}^2+\frac{1}{2}m_2\,{{v_2}'}^2+Q$$

Conservation of momentum:

$$m_1\,v_1+m_2\,v_2=m_1\,v_1'+m_2\,v_2'$$

For *completely inelastic shocks* is

$$v_1'=v_2'=v'\quad .$$

Both bodies move with the same velocity $\vec{v}'$ after the impact.

$$v'=\frac{m_1\,v_1+m_2\,v_2}{m_1+m_2}$$
