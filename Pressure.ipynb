### 7.1.2 Density

The density $\rho$ of a body is defined as its mass per volume:

$$\rho=\frac{m}{V}\quad ,$$ 

the SI unit $[\rho]=$kg/m$^3$.

### 7.1.3 Pressure in liquids

The pressure $p$ is the force per unit area. If a homogeneous force $F$ acts vertically on the surface $A$, the pressure is 

$$p=\frac{F}{A}$$


$$\left[p\right]=\frac{\mathrm{N}}{\mathrm{m}^2}=\mathrm{Pa}$$

The atmospheric normal pressure is $1013\,\mathrm{hPa}=1\,\mathrm{atm}=760\,\mathrm{Torr}$.

### 7.1.4 Pascal's law

The pressure in a fluid at rest (not including gravity) is the same at all points and in all directions.

A force $\vec{F}$ acts on a vessel wall, the force of the vessel wall on the fluid is $-\vec{F}$.

<img src="figs/9n.1_1.png" width=200px>

In general, the pressure can be defined as the quantity per surface element $\mathrm{d}\vec{A}$ that generates a force $\mathrm{d}\vec{F}$ perpendicular to the surface.

$$\mathrm{d}\vec{F}=-p{\mathrm{d}\vec{A}}=-p \,\vec{n}\,\mathrm{d}A$$

The force due to the pressure of the fluid is always perpendicular to each surface and acts on the surface with which the fluid is in contact.

### 7.1.5 Hydrostatic pressure 

We now calculate the change in pressure exerted by gravity in a fluid. The density can be assumed to be constant to a good approximation and we obtain the gravitational pressure 

<img src="figs/9.1_2.png" width=200px>

Gravitational pressure:

$$p_S=\frac{F}{A}=\frac{mg}{A}=\frac{\rho_{Fl} A g h}{A}=\rho_{Fl} g h$$

Hydrostatic pressure:

$$p_H=p_S+P_{\mathrm{Luft}}=p_{ges}(h)$$

### 7.1.6 U-tube manometer

The knowledge that the pressure difference is proportional to the depth $h$ in a liquid can be used to measure unknown pressures. The figure below shows a simple pressure gauge, the open manometer, which is also called a U-tube manometer because of its shape.

### 7.1.7 Buoyancy

* Archimedes' principle

<img src="figs/9.1_3.png" width=220px>

$$\vec{F}_1=-m_{Fl}^{(1)}\,g\,\vec{e}_x$$

$$\vec{F}_2=+m_{Fl}^{(2)}\,g\,\vec{e}_x$$

$$F_{A}=\vec{F}_1+\vec{F}_2=g \, \rho_{Fl}\,A\,(h_2-h_1)\,\vec{e}_x$$

The volume of the body is $A(h_2-h_1)=V_{\mathrm{body}}$.

And the mass of the displaced liquid is $m_{Fl}=\rho_{Fl}\,V_{\mathrm{body}}$.

$$\vec{F}_{A}=\rho_{Fl}\, g\, V_{\mathrm{KÃ¶rper}}\,\vec{e}_x$$.

*Archimedes' principle* The buoyant force experienced by a body immersed in a fluid is equal to the weight of the fluid displaced by this body.


The apparent weight $F'_G$ of a body immersed in a fluid is the difference between the weight $|\vec{F}_G|$ and the buoyant force $|\vec{F}_A|$:

$$F'_G=|\vec{F}_G|-|\vec{F}_A|$$


From this, the following cases can be derived for $F'_G$:

* $F'_G>0$ the body sinks
* $F'_G=0$ the body floats
* $F'_G<0$ the body floats

## 7.2 Phenomena at interfaces
### 7.2.1 Surface tension

Cohesive forces: between the molecules inside the liquid

$$ \vec{F}_{ij}=0$$

<img src="figs/9n.2_1.png" width=350px>

Adhesive forces act at the edges of the liquid, i.e. towards solid and gaseous substances or other liquids. These can be stronger or weaker than the cohesive force. In the case of gases, the adhesion is significantly lower and a resulting force acts on the liquid molecules at the liquid-gas interface into the liquid.

$$ \vec{F}_{ij}=\vec{F}_{res}\ne0$$

#### Surface tension
* To increase the surface area of a liquid, molecules must be brought to the surface => work against $\vec{F}_{res}$.
* Increase of the surface energy:

  $$\sigma=\frac{\Delta W}{\Delta A}$$
  
  $\sigma$: specific surface energy
  
  $\Delta W$: Work
  
  $\Delta A$: Change in surface area

* Determination of $\sigma$ using the stirrup method

<img src="figs/9.2_2.png" width=300px>  

$$\Delta W=F\,\Delta s$$

$$\sigma=\frac{F\,\Delta s}{2\,\Delta s\,l}=\frac{F}{2\,l}$$

* Pressure in soap bubbles

<img src="figs/9.2_3.png" width=350px>  

 $p$ is the pressure inside the bubble, $p_0$ the external air pressure, $A$ the surface of the bubble. 
 
 The bubble tries to reach the state of lowest surface energy $\rightarrow$ reduction of the surface area. 
 
 Inside, there is an overpressure $\Delta p_{SB}=p-p_0$. This leads to a force directed radially outwards 
 
 $$F_p=\Delta p_{SB}\,A$$
 
 We now want to determine this pressure difference:
 
 The energy gain $\Delta W$ when the surface is reduced is
 
 \begin{eqnarray}
 \Delta W=\sigma \, \Delta A&=&2\sigma\,4\pi\left[\left(r+\Delta r\right)^2-r^2\right]\\
 &=&2\sigma\,4\pi\left[2\,r\,\Delta r+\Delta r^2\right]\\
 &\approx&4\sigma\,4\pi(\,r\,\Delta r)\quad ,
 \end{eqnarray}
 since $\Delta r^2$ can be neglected for $\Delta r\rightarrow 0$. The factor 2 comes from the two surfaces of the bubbles.
 
 The work to be done against $\Delta p$
 
 $$\Delta W_A=F_P\,\Delta r=\Delta p_{SB}\, \underbrace{4\,\pi\,r^2}_A\,\Delta r$$
 
 Equilibrium prevails when 
 
 \begin{eqnarray}
 \Delta W &=& \Delta W_A\\
 4\sigma &=& \Delta p_{SB} \, r\\
 \Delta p_{SB} &=& \frac{4\sigma}{r}
 \end{eqnarray}
 
 The pressure difference $\Delta p$ increases the smaller the soap bubble becomes.
 
 A very similar derivation applies to bubbles in liquid, except that only *one* surface has to be taken into account. We therefore obtain
 
 $$\Delta p_B = \frac{2\sigma}{r}\quad .$$
 
### 7.2.2 Interfacial tension & wetting
 
* Wetting

 <img src="figs/9n.2_4.png" width=350px>  
 
 We speak of wetting when the adhesion forces are greater than the cohesion forces. A distinction is made between the forces for the three phases: solid (3), liquid (2) and gaseous (1). If the droplet assumes a fixed position with the contact angle $\alpha$, the forces must add up to zero. The equilibrium of forces results in

 $$\sigma_{13}=\sigma_{23} + \sigma_{12}\cos\alpha$$
 
 If the contact angle $\alpha<=\pi/2$, the liquid wets the surface.
 
 <img src="figs/9n.2_5.png" width=270px> 
 
 If the contact angle is $\alpha>\pi/2$, there is no wetting. The liquid contracts to form a drop. The cohesive forces are therefore greater than the adhesive forces.
 
 In the case of water, one speaks of hydrophilic when the surface is wetted and hydrophobic when it is not wetted. 
 
 
### 7.2.3 Capillarity
 
 In capillaries (thin tubes with a radius $r$ ) that are wetted, the liquid level rises. In non-wetting capillaries (e.g. glass and mercury), the liquid level falls or a capillary depression.
 
 <img src="figs/9n.2_6.png" width=500px>  
 
 The surface tension and the weight force balance each other out in such a capillary. The force $F_\sigma$ resulting from the surface tension is equal to the weight force $F_g$, i.e. $F_\sigma=F_g$. 
 
 $$F_\sigma=\sigma\,2\pi\,r$$
 
 $$F_g=m\,g\,h=\rho_{Fl}\pi\,r^2\,h$$
 
 Converting to $h$ results in a climb height of
 
 $$h=\frac{2\sigma_{12}\cos\alpha}{\rho_{Fl} g r}$$
 
 Here we have still used the relationship between surface tension $\sigma_{12}$ and contact angle.
 
The rise height can be positive and negative, depending on whether the liquid wets the surface or not. The rise height increases for small capillaries. This is important for plants and trees, for example, which use capillaries for liquid transport.
