### 7.1.1 Free displaceability

* Ideal fluids (gases and liquids)
  * No friction
  * No interfacial interaction between two fluids

### 7.1.2 Density

The density $\rho$ of a body is defined as its mass per volume:

$$\rho=\frac{m}{V}\quad ,$$ 

the SI unit is $[\rho]=$kg/m$^3$.

Examples of the density of solids, liquids and gases:

|Substance|density in kg/m$^3$|
|---|---|
| Iron | $7.8\cdot 10^3$ |
| Lead | $11.3\cdot 10^3$|
| wood | $0.3-0.9 \cdot 10^3$|
| Ice | $0.917\cdot 10^3$|
| Water (4$^\circ$C)|$1.00\cdot 10^3$|
| Mercury | $13.6\cdot 10^3$|
| Gasoline | $0.68\cdot 10^3$|
|Air | $1.29$|
|Helium | $0.179$|
|Water vapor (100$^\circ$C) | $0.598$|

The density in liquids is almost constant, but in gases it can change due to a force acting on the gas. We call this behavior compressible. Liquids in contrast are nearly incompressible.

### 7.1.3 Pressure in liquids

The pressure $p$ is the force per unit area. If a homogeneous force $F$ acts normal on a surface $A$, the pressure is defined as

$$p=\frac{F}{A}$$


$$\left[p\right]=\frac{\mathrm{N}}{\mathrm{m}^2}=\mathrm{Pa}$$

The atmospheric normal pressure is $1013\,\mathrm{hPa}=1\,\mathrm{atm}=760\,\mathrm{Torr}$.

### 7.1.4 Pascal's law

The pressure in a fluid at rest (and in absence of gravity) is the same at all points and in all directions.

A force $\vec{F}$ acts on a vessel wall, the force of the vessel wall on the fluid is $-\vec{F}$.

<img src="figs/9n.1_1.png" width=200px>

In general, the pressure can be defined as the quantity per surface element $\mathrm{d}\vec{A}$ that generates a force $\mathrm{d}\vec{F}$ perpendicular to the surface.

$$\mathrm{d}\vec{F}=-p{\mathrm{d}\vec{A}}=-p \,\vec{n}\,\mathrm{d}A$$

The force due to the pressure of the fluid is always perpendicular to a surface and acts on the surface with which the fluid is in contact.

### 7.1.5 Hydrostatic pressure 

We now calculate the change in pressure exerted by gravity on a fluid particle. The density can be assumed constant in good approximation and we obtain the hydrostatic pressure 

<img src="figs/9n.1_2.png" width=200px>

Hydrostatic pressure:

$$p_S(h)=\frac{F}{A}=\frac{mg}{A}=\frac{\rho_{fl} A g h}{A}=\rho_{fl} g h$$

Total pressure:

$$p_{tot}(h)=p_S(h)+p_{\mathrm{air}}$$

### 7.1.6 U-tube manometer

The knowledge that the pressure difference is proportional to the depth $h$ in a liquid can be used to measure an unknown pressures. The figure below shows a simple pressure gauge, the open manometer, which is also called a U-tube manometer because of its shape.

If the liquid in the U-tube is in mechanical equilibrium, the unknown pressure $p$ is equal to $p_{\mathrm{air}}$ and the pressure of the liquid column with length $h$:

$$p=p_\mathrm{air}+\rho_{fl}\,g\, h$$

### 7.1.7 Buoyancy

* Archimedes' principle

<img src="figs/9.1_3.png" width=220px>

$$\vec{F}_1=-m_{Fl}^{(1)}\,g\,\vec{e}_x$$

$$\vec{F}_2=+m_{Fl}^{(2)}\,g\,\vec{e}_x$$

$$F_{b}=\vec{F}_1+\vec{F}_2=g \, \rho_{Fl}\,A\,(h_2-h_1)\,\vec{e}_x$$

The volume of the body is $A(h_2-h_1)=V_{\mathrm{body}}$.

And the mass of the displaced liquid is $m_{Fl}=\rho_{Fl}\,V_{\mathrm{body}}$.

$$\vec{F}_{b}=\rho_{Fl}\, g\, V_{\mathrm{body}}\,\vec{e}_x$$.

*Archimedes' principle* The buoyant force experienced by a body immersed in a fluid is equal to the weight of the fluid displaced by this body.


The apparent weight $F'_G$ of a body immersed in a fluid is the difference between the weight $|\vec{F}_G|$ and the buoyant force $|\vec{F}_A|$:

$$F'_G=|\vec{F}_G|-|\vec{F}_A|$$


From this, the following cases can be derived for $F'_G$:

* $F'_G>0$ the body sinks
* $F'_G=0$ no net force and the body is neutrally buoyant
* $F'_G<0$ the body floats

## 7.2 Phenomena at interfaces
### 7.2.1 Surface tension

Cohesive forces: between the molecules inside the liquid

$$ \vec{F}_{ij}=0$$

<img src="figs/9n.2_1.png" width=350px>

Adhesive forces act at the edges of the liquid, i.e. towards solid and gaseous substances or other liquids. These can be stronger or weaker than the cohesive force. In the case of gases, the adhesion is significantly lower and a resulting force acts on the liquid molecules at the liquid-gas interface into the liquid.

$$ \vec{F}_{ij}=\vec{F}_{res}\ne0$$

#### Surface tension
* To increase the surface area of a liquid, molecules must be brought to the surface => work against $\vec{F}_{res}$.
* Increase of the surface energy:

  $$\sigma=\frac{\Delta W}{\Delta A}$$
  
  $\sigma$: specific surface energy
  
  $\Delta W$: Work
  
  $\Delta A$: Change in surface area

* Determination of the coefficient of surface tension $\sigma$ using the anchor-ring method

<img src="figs/9.2_2.png" width=300px>  

$$\Delta W=F\,\Delta s$$

$$\sigma=\frac{F\,\Delta s}{2\,\Delta s\,l}=\frac{F}{2\,l}$$

* Pressure in the soap bubbles

<img src="figs/9.2_3.png" width=350px>  

 $p$ is the pressure inside the bubble, $p_0$ the external air pressure, $A$ the surface of the bubble. 
 
 The bubble tries to reach the state of lowest surface energy $\rightarrow$ reduction of the surface area. 
 
 Inside, there is an overpressure $\Delta p_{SB}=p-p_0$. This leads to a force directed radially outwards 
 
 $$F_p=\Delta p_{SB}\,A$$
 
 We now want to determine this pressure difference:
 
 The energy gain $\Delta W$ when the surface area is reduced is
 
 \begin{eqnarray}
 \Delta W=\sigma \, \Delta A&=&2\sigma\,4\pi\left[\left(r+\Delta r\right)^2-r^2\right]\\
 &=&2\sigma\,4\pi\left[2\,r\,\Delta r+\Delta r^2\right]\\
 &\approx&4\sigma\,4\pi(\,r\,\Delta r)\quad ,
 \end{eqnarray}
 since $\Delta r^2$ can be neglected for $\Delta r\rightarrow 0$. The factor 2 comes from the two surfaces of the bubbles.
 
 The work to be done against $\Delta p$
 
 $$\Delta W_A=F_P\,\Delta r=\Delta p_{SB}\, \underbrace{4\,\pi\,r^2}_A\,\Delta r$$
 
 Equilibrium prevails when 
 
 \begin{eqnarray}
 \Delta W &=& \Delta W_A\\
 4\sigma &=& \Delta p_{SB} \, r\\
 \Delta p_{SB} &=& \frac{4\sigma}{r}
 \end{eqnarray}
 
 The pressure difference $\Delta p$ increases the smaller the soap bubble becomes.
 
 A very similar derivation applies to bubbles in liquid, except that only *one* surface has to be taken into account. We therefore obtain
 
 $$\Delta p_B = \frac{2\sigma}{r}\quad .$$
 
### 7.2.2 Interfacial tension & wetting
 
* Wetting

 <img src="figs/9n.2_4.png" width=350px>  
 
 We speak of wetting when the adhesion forces are greater than the cohesion forces. A distinction is made between the forces for the three phases: solid (3), liquid (2) and gaseous (1). If the droplet assumes a fixed position with the contact angle $\alpha$, the forces must add up to zero. The equilibrium of forces results in

 $$\sigma_{13}=\sigma_{23} + \sigma_{12}\cos\alpha$$
 
 If the contact angle $\alpha<=\pi/2$, the liquid wets the surface.
 
 <img src="figs/9n.2_5.png" width=270px> 
 
 If the contact angle is $\alpha>\pi/2$, there is no wetting. The liquid contracts to form a drop. The cohesive forces are therefore greater than the adhesive forces.
 
 In the case of water, one speaks of hydrophilic when the surface is wetted and hydrophobic when it is not wetted. 
 
 
### 7.2.3 Capillarity
 
 In capillaries (thin tubes with a radius $r$ ) that are wetted, the liquid level rises. In non-wetting capillaries (e.g. glass and mercury), the liquid level falls or a capillary depression.
 
 <img src="figs/9n.2_6.png" width=500px>  
 
 The surface tension and the weight force balance each other out in such a capillary. The force $F_\sigma$ resulting from the surface tension is equal to the weight force $F_g$, i.e. $F_\sigma=F_g$. 
 
 $$F_\sigma=\sigma\,2\pi\,r$$
 
 $$F_g=m\,g\,h=\rho_{Fl}\pi\,r^2\,h$$
 
 Converting to $h$ results in a climb height of
 
 $$h=\frac{2\sigma_{12}\cos\alpha}{\rho_{Fl} g r}$$
 
 Here we have still used the relationship between surface tension $\sigma_{12}$ and contact angle.
 
The rise height can be positive and negative, depending on whether the liquid wets the surface or not. The rise height increases for small capillaries. This is important for plants and trees, for example, which use capillaries for liquid transport.

## 7.3 Flowing liquids and gases (hydro- and aerodynamics)

### 7.3.1 Basic concepts

* Dynamics: 

  Sum of all momentum of the molecules 
  
  $$\sum \vec{p}_i\ne 0$$
  
  $\rightarrow$ flowing liquids/gases
  
* relative motion in relation to a fixed laboratory coordinate system

* motion is slow to the speed of sound (propagation speed of pressure changes)

* thermal motion is not taken into account
* Consideration of volume elements

  $$\Delta V = \frac{\Delta m}{\rho}$$

* Movement due to 
  * Pressure forces
  * gravitational force
  * frictional forces
  
<img src="figs/10n.1_1.png" width=420px>


### 7.3.2 Continuity equation

* Flow through a surface $\vec{A}$: the quantity $\vec{v}\cdot\vec{A}$ indicates the volume flow through the surface $\vec{A}$. The area is a vector, the magnitude indicates the size of the area and the vector the direction. 

<img src="figs/10.1_2.png" width=300px>

The volume flow $Q$ is 

$$Q=\vec{v}\cdot\vec{A}=\underbrace{|\vec{v}|\,|\vec{A}|\,\cos(\vec{v},\vec{A})}_{A_\perp}=|\vec{v}|\,A_\perp= v\,A$$


* Continuity equation

<img src="figs/10.1_3.png" width=420px>

$$\Delta V_1=A_1\Delta x_1\quad\quad\quad\Delta V_2=A_2\Delta x_2$$

* Conservation of the mass flow through the area $A_1$ or $A_2$

* The mass flow is

$$\frac{\Delta m}{\Delta t}$$

This means that $$\Delta m_1=\rho_1\,\Delta V_1$$
and $$\Delta m_2=\rho_2\,\Delta V_2$$


$$\frac{\Delta m_1}{\Delta t}=\frac{\Delta m_2}{\Delta t}$$

$$\rho_1\,A_1\,\frac{\Delta x_1}{\Delta t}=\rho_2\,A_2\,\frac{\Delta x_2}{\Delta t}$$

Border crossing $\Delta t \rightarrow 0$:

$$\rho_1\,A_1\,\frac{\mathrm{d} x_1}{\mathrm{d} t}=\rho_2\,A_2\,\frac{\mathrm{d} x_2}{\mathrm{d}t}$$

and with the definition for the oscillation we obtain the continuity equation for the mass flow:

$$\rho_1\,A_1\,v_1=\rho_2\,A_2\,v_2$$

If the density does not change, e.g. in incompressible liquids/gases, $\rho_1=\rho_2$ and the following applies

$$A_1\,v_1=A_2\,v_2=\mathrm{konst.}$$

### 7.3.3 Bernoulli equation


The force exerted on the segment by the fluid to the left of the blue portion in below Figure has a magnitude $F_1 = p_1 A_1$. During a time interval $\Delta t%$, the point of application  of  this  force  travels a distance of $\Delta x_1$,  as  the  blue  portion of fluid enters the section of pipe past point 1. The work done by this force is $W_1= F_1 \Delta x_1 =p_1 A_1 \Delta x_1 = p_1 V$, where $V$ is the volume  of the  blue  portion  of  fluid  passing  point  1.  

In  a  similar  manner, the work done by the segment by the fluid to the right of the segment in the  same  time  interval  $\Delta t$ is  $W_2 = - p_2 A_2 \Delta x_2 = p_2 V$,  where  $V$  is  the same volume as on the left. The negative sign comes from the fact that the work is done by the volume segment. 

<img src="figs/10.1_4.png" width=600px>

The  net  work  done  on  the  segment  by  these  forces  in  the  time  interval $\Delta t$ is then 

$$
W=(p_1-p_2) V
$$

This work changes the kinetic and potential energy of the fluid particles, 
thus 

$$
\Delta E=E_2-E_1=\frac{1}{2}\rho V v_2^2 + \rho V g y_2 -  \frac{1}{2}\rho V v_1^2 - \rho V g y_1=(p_1-p_2) V
$$

Now moving the energies and work at point 1 and point 2 on each side of the equation and dividing by the volume $V$ gives us:

$$
p_1+\frac{1}{2}\rho v_1^2 + \rho g y_1 = p_2+\frac{1}{2}\rho v_2^2 + \rho g y_2
$$

This equation is termed the Bernoulli equation, named after the Swiss physicist Daniel Bernoulli, who published the important book *Hydrodynamica* in 1738. If your are interested and can read Latin, here is the [link](https://www.e-rara.ch/zut/doi/10.3931/e-rara-3911) (chapter 12, or see [here](https://www.17centurymaths.com/contents/dbernoullicontents.htm) for the translation into English).

Bernoulli equation can be applied along a streamline in a steady (time independent) flow at two or more points. 


### Applications of Bernoulli Equation

#### Venturi Tube

<img src="figs/10.1_5.png" width=400px>


The  horizontal  constricted  pipe  as illustrated  in the  figure is named Venturi tube. By measuring the pressure of the flow at to points the liquid velocity can be be determined.

We apply Bernoulli equation for $y_1=y_2=0$, as the center streamline is on the same height.

$$
p_1 + \frac{1}{2} \rho v_1^2 = p_1 + \frac{1}{2} \rho v_1^2
$$

The ratio of the velocities because of conservation of mass is

$$v_1=\frac{A_2}{A_1}v_2\quad .$$

Insert this expression into Bernoulli equation and solve for $v_2$ gives us

$$
v_2=\sqrt{\frac{2(p_1-p_2)}{\rho(A_1^2-A_2^2)}}\quad .
$$


The pressure difference $\Delta p= p_1-p_2$ can be conveniently measured with a U-tube manomoeter.

#### Toricelli's law

An enclosed tank with a constant presure $p$ is emptied into the surrounding with constant pressure $p_0$ through a nozzle with an opening area of $A_1$. $A_1$ is much smaller than $A_2$, thus we can neglect the velocity at the free surface, i.e. $v_2\approx 0$.

<img src="figs/10.1_6.png" width=300px>

Applying Bernoulli between points 1 and 2 we obtain

$$
p_0+\frac{1}{2}\rho v_1^2 +\rho g y_1 =p+\rho g y_2 \quad .
$$

This can be solved for $v_1$ and we obtain

$$v_1=\sqrt{\frac{2(p-p_0)}{\rho}+2 g h}\quad .
$$
