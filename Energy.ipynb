# Mechanical and Chemical Equilibrium in the Living Cell

Living systems consume and liberate energy through chemical transformations. How cells manipulate and store energy. For this we need to introduce the concept of an equilibrium state. Most cells are close to chemical and mechanical equilibrium.

Four kinds of energy are important in biology: chemical energy, mechanical energy, electromagnetic energy, and thermal energy. Living organism can convert these energies between each other, except for thermal energy. Thermal energy is kind of "dead", it can not be converted to do useful work. Example for energy conversion, sunlight is converted to chemical energy through the process of photosynthesis. Fireflies convert chemical energy into light. Energy can be stored mostly in chemical form. Energy storing molecules are for example ATP is used to pump molecules across the membrane, to drive polymeric motors for cellular motility and to power the motors for muscle contractions.

Biological system is operating close to equilibrium, any small devitation will allow the system to return back to its original state. This is described by energy minimization (free energy minimization). 

In biological systems on the cellular and sub-cellular level thermal and deterministic forces are similarly strong. Thermal forces are forces exterted on macromolecular structures as a result of thermal motion of all molecules (incl. water).  This thermal jostling must be accounted for in biological systems. 

The most famous motion is Brownian motion. Small particles of about $1\,\mu$m and smaller, fluorescently labled molecules and macromolecules undergo excursions that appear completely random. This reminds us that we have to add to any deterministic force always the forces from constant collisions with molecules from the surrounding medium.

## Energy minimization

The DNA sequence is translated into a amino acid sequence. The function of this resulting protein or peptide is strongly dependent on the folding of the amino acid sequence. Thus in what possible configuration does the very long sequence fold?

This can be answered with physics. The mechanical equilibrium state is obtained by minimizing the potential energy of the physical system. At equilibrium no unbalanced forces exits. We can write this usinhg the mathematical language 

$$\sum_i \mathbf{F}_i = 0\quad ,$$

where $mathbf{F}_i$ is the $i$-th force acting on the system. The force is written here as a vector, that is why we use the bold character $\mathbf{F}$. Other ways to write a vector is using $\vec{F}$ or sometimes $\underline{F}$.
