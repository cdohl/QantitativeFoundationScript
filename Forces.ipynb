## 3.1 Circular motion

### Plane polar coordinates

Radius $r=\left|\vec{r}\right|$

Angle $\varphi$


### Circular path

$\left|\vec{r}\right|=r=\mathrm{konst.}$

Angle: $\varphi=\varphi(t)$

$$x(t)=r\,\cos \left(\varphi(t)\right)$$

$$y(t)=r\,\sin \left(\varphi(t)\right)$$

* Angular velocity

  $$\omega=\frac{\mathrm{d}\varphi}{\mathrm{d}t}$$
  
  $$\left[\varphi\right]=1\,\mathrm{rad}$$
  
  $$\left[\omega\right]=\frac{\mathrm{rad}}{\mathrm{s}}\stackrel{\frown}{=}\frac{1}{\mathrm{s}}$$
  
 
* Defintion of angular acceleration in general, yet we will not deal further with it

  $$\alpha=\frac{\mathrm{d}\omega}{\mathrm{d}t}=\dot{\omega}=\frac{\mathrm{d}^2\varphi}{\mathrm{d} t^2}=\ddot{\varphi}$$
  
### Circular motion

* uniform circular motion: $\omega=\mathrm{const}$ and $\alpha=0$
* non-uniform circular motion:  $\omega\ne\mathrm{const}$ and $\alpha\ne0$


* Example for uniform circular motion
  \begin{eqnarray}
  \alpha &=& \frac{\mathrm{d}\omega}{\mathrm{d}t} &=& 0\\
  \int \mathrm{d}\omega &=& \int 0 \, \mathrm{d}t\\
  \omega &=& C_1=\mathrm{const.}
  \end{eqnarray}
  
  The time dependent angle is

  \begin{eqnarray}
  \omega&=&\frac{\mathrm{d}\varphi}{\mathrm{d}t}\\
  \mathrm{d}\varphi&=&\omega\,\mathrm{d}t\\
  \int\mathrm{d}\varphi&=&\int \omega\,\mathrm{d}t\\
  \varphi(t)&=&\omega\,t + \varphi_0 
  \end{eqnarray}
  
  $\varphi_0$ is the initial angle, i.e. $\varphi_0=\varphi(t=0)$.
  
  ⇒ The period for one full turn $T$
  
  $$\omega=\frac{2\,\pi}{T}$$
  
  ⇒ frequency (number of turns per time )
  
  $$f=\frac{1}{T}=\frac{\omega}{2\,\pi}$$
  
  Connection between angular velocity and frequency 
  
  $$\omega=2\,\pi\,f$$  
  

```python
%matplotlib inline
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import matplotlib.patches as patches
mpl.rcParams['font.size'] = 15

R=1
alpha=np.linspace(0,2*np.pi,200)



def update_plot(phi):
    
    phirad=phi*np.pi/180.
    plt.figure(1, figsize=(6, 6))
    ax=plt.gca()
    plt.plot(R*np.cos(alpha),R*np.sin(alpha))
    ax.axhline(color='k',linewidth=1);ax.axvline(color='k',linewidth=1)
    ax.text(R*1.3,-0.1,r'$x$');ax.text(-0.1,R*1.3,r'$y$')
    ax.set_xticks([]);ax.set_yticks([])
    ax.spines['top'].set_visible(False);ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False);ax.spines['left'].set_visible(False)
    
    #velocity vector
    ax.arrow(R*np.cos(phirad), R*np.sin(phirad), -.7*np.sin(phirad), .7*np.cos(phirad), lw=1., fc='k', ec='k', head_width=.05, head_length=.05, length_includes_head=True)
    ax.text(R*np.cos(phirad)-.35*np.sin(phirad),R*np.sin(phirad)+.35*np.cos(phirad),r'$\vec{v}$',fontsize=18)
    #angle
    arc = patches.Arc((0.,0), R, R, theta1=0., theta2=phi,linewidth=2.)
    ax.add_patch(arc)
    ax.text(R*0.3,0.1,r'$\varphi$')
    plt.plot([0,R*np.cos(phirad)],[0,R*np.sin(phirad)],'k')
    plt.plot([R*np.cos(phirad),R*np.cos(phirad)],[0,R*np.sin(phirad)],'k--')
    plt.plot([0,R*np.cos(phirad)],[R*np.sin(phirad),R*np.sin(phirad)],'k--')
    ax.text(.5*R*np.cos(phirad)-.05*np.sin(phirad),.5*R*np.sin(phirad)+.05*np.cos(phirad),r'$\vec{r}$')
    #axis limits
    ax.set_xlim([-1.3*R, 1.3*R])
    ax.set_ylim([-1.3*R, 1.3*R])
    plt.show();


    
tv = widgets.FloatSlider(value=30., min=0., max =360., step=15.,description=r'$\varphi$:');
widgets.interactive(update_plot, phi=tv)
```

#### Every circular motion is an accelerated motion!

Every circular movement is an accelerated movement, as at least the direction of $\vec{v}$ is constantly changing.  

<img src="figs/3.5_1.png" width="300">

You can see this clearly if you compare the two vectors $\vec{v}_1$ and $\vec{v}_2$.

<img src="figs/3.5_2.png" width="100">


$$\vec{v}_2=\vec{v}_1+\Delta \vec{v}$$

Let us first have a look at the the average acceleration for an object moving uniformly (with constant angular velocity). Then the average acceleration is the magnitude of the change of $\vec{\Delta v}$ over time or 

$$|\vec{a_{avg}}|=\frac{\Delta v}{\Delta t}\quad .$$

Looking at both triangles composed of $\vec{r_1}$, $\vec{r_2}$, and $\vec{\Delta r}$ versus $\vec{v_1}$, $\vec{v_2}$, and $\vec{\Delta v}$ in the two figures above. They are similar, i.e. the angles in both triangles are the same. Thus we can write 

$$\frac{|\vec{\Delta v}|}{v}=\frac{|\vec{\Delta r}|}{r}\quad .$$

With this we can express $|\vec{\Delta v}|$ as

$$|\vec{\Delta v}|=\frac{v |\vec{\Delta r}|}{r}\quad .$$

Inserting this into the expression for the average acceleration we obtain

$$|\vec{a_{avg}}|= \frac{v |\vec{\Delta r}|}{r\,\Delta t}$$


The instantaneous acceleration can be obtained by taking the limit $\Delta t\rightarrow 0$ then the we obtain

$$|\vec{a_{c}}|=a_c= \frac{v^2 }{r}\quad .$$

This instantaneous acceleration is termed centripetal acceleration. It acts towards the center of the circle.

## 3.2 Dynamics of the point masses
### 3.2.1 Concept of force and Newton's laws

#### Newton's 1st law

<b>In an inertial frame, a body moves in a straight line at a constant speed if no force acts on the body.</b>


An inertial system is a non-accelerated coordinate system. *Is the lecture hall an inertial frame?*

Let's look at the radial acceleration on earth:

<img src="figs/4.1_1.png" width=250px>

If we use the earth's radius of approx. $R_e\approx 6400\,$km and the angular velocity of $\omega=2\,\pi/(24*3600)\,$s$^{-1}$, the radial acceleration at the equator is a maximum of $a_r=0.034\,$m/s$^2$.
Although this is significantly less than the gravitational acceleration, it is still an accelerated coordinate system. Although the lecture hall is therefore not an exact inertial frame, it is a good approximation.


#### Newton's 2nd law

Take a spring and extend the spring by a fixed distance. Then we add a mass $m_1$ to the spring. The spring pulls on the mass $m_1$ and accelerates it to $a_1$. If we repeat the experiment and attach a different mass $m_2$ starting with the same extension of the spring, then the spring pulls on this mass $m_2$ with a different acceleration $a_2$. 

<img src="figs/4.1_2.png" width=350px>

From experiments we know that the product of mass and acceleration is the same for both masses for the same extension of the spring, i.e. 

$$m_1\,a_1=m_2\,a_2\quad.$$

The product of $m$ and $a$ is the definition of force. Newton's 2nd law states:


<b> A force on a body accelerates it in the direction of the force and has the magnitude $m\,a$.</b>

$$\vec{F}=m\,\vec{a}$$

Pease note that this only applies to bodies with a constant mass. The unit of $F$ is

$$\left[F\right]=\mathrm{kg}\,\frac{\mathrm{m}}{\mathrm{s}^2}=\mathrm{N}$$ 

N is for Newton.

The gravitational force is thus

$$\vec{F}_g=m\,\vec{g}\quad ,$$

where $\vec{g}$ is the acceleration due to gravity. If you hold a blue ball in your hand as shown in the picture below, the ball does not move. Therefore, the force on the ball must be 0. In addition to the gravitational force $\vec{F}_g$, also the force of the hand $\vec{F}_{\mathrm{Hand}}$ acts on the ball.

<img src="figs/4.1_3.png" width=300px>

The total force is the sum of the forces and therefore 0:

$$\vec{F}_g+\vec{F}_{\mathrm{Hand}}=0\quad$$

and thus we have

$$\vec{F}_{\mathrm{Hand}}=-m\vec{g}\quad .$$

#### Example balance of forces

The mass $m$ in below figure is at rest. 

<img src="figs/4.1_4.png" width=400px>

As a consequence the the acceleration of the mass is 0 and thus

$$\sum_i\vec{F}_i=m\vec{a}=0$$

which means that the sum of the forces acting on the object is 0, or
$$\vec{F}_1+\vec{T}_2+\vec{T}_3 =0$$

This holds for direction in $x$ and in $y$, we can thus write equations for these three forces being 0 in sum seperately. 

Let us start with the $x$-direction:

$$T_{2x}+T_{3x}=0$$ 

We can rewrite this equation in terms of magnitude and angle such as 

$$-T_2\sin\alpha+T_3\sin\beta=0\tag{3.1}\quad .$$

Similarily for the $y$-direction:

$$-m\,g+T_2\cos\alpha+T_2\cos\beta=0\tag{3.2}$$

Here we used these identities: $\sin 30^\circ=\cos 60^\circ=\frac{1}{2}$ und $\cos 30^\circ=\sin 60^\circ=\frac{\sqrt{3}}{2}$.


Let us rewrite Eq. (3.1) 

$$T_2=\sqrt{3}T_3$$ 

and insert into Eq. (3.2), then solve for $T_3$:

$$T_3=\frac{m\,g}{2}$$

thus we have 
$$T_2=\frac{\sqrt{3}}{2}m\,g$$


This can obtained even faster. At rest the sum of the vectors $\vec{T}_2$ and $\vec{T}_3$ must be identical to $-\vec{F}_1$. And because of $\alpha+\beta=90^\circ$ we can use Pythagoras as such:

$$T_2^2+T_3^2=m\,g$$

<img src="figs/4.1_5.png" width=150px>

#### 3rd Newton's law

<b>If a body acts with a force on a second body, the second body acts with the same force but in the opposite direction. Every force has an opposing force: action = -reaction.</b>
    
*Example:*

Let us have a look at the force and its transduction onto two bodies. $\vec{F}$ is acting on body 1, und then onto body 2. Let the masses be $m_1=5\,$kg and $m_2=15\,$kg, and the force $F=20\,$N acting to the right (positive direction).

<img src="figs/4.1_6.png" width=250px>

Both bodies are in contact, thus they have the same acceleration

$$F=m\,a=(m_1+m_2)\,a$$

Thus the accelearation is $a=1\,$ms$^{-2}$. 

<img src="figs/4.1_7.png" width=180px>

Let us now focus on body 2 with mass $m_2$. On this body the contact force from mass $m_1$ onto mass $m_2$ is acting, which we call $F_{12}$. This must be 

$$F_{12}=m_2\,a=15\,\mathrm{N}\quad .$$

Now we draw the force acting on body 1 with mass $m_1$:

<img src="figs/4.1_8.png" width=200px>

The force $F$ acts towards the right and the contact force $F_{21}$, from mass $m_2$ onto mass $m_1$ to the left. Thus

$$F+F_{21}=m_1\,a=5\,\mathrm{N}$$

$$20\,\mathrm{N}+F_{21}=5\,\mathrm{N}$$

$F_{21}=-15\,$N. We thus find that the action force is the negative of the reaction force, i.e. $F_{12}=-F_{21}$.



*Consequences of Newton's 3rd law*

The very same holds for all forces, for example if an apple drops from the the tree onto Earth:


<img src="figs/4.1_9.png" width=250px>

Let the mass of the apple be $m_{\mathrm{Apfel}}=0.5\,$kg, the mass of Earth $m_{\mathrm{Erde}}=6\cdot 10^{26}\,$kg und the apple is dropping from a height of $2\,$m.

Gravity is acting on the apple with $F=m\,g$. According to 3rd Newton's law a opposite force of $-F$ must be acting. Thus we may ask the question, how much does Earth move towards the apple if it is dropped from a height $h$?


$$F=m\,g=0.5\,\mathrm{kg}\,10\,\mathrm{m}\mathrm{s}^{-2}=5\,\mathrm{N}$$

The duration of the fall can be calculated as:

$$h=\frac{1}{2}\,g t^2\quad .$$

Let solve for $t$ and we obtain a fall time of about $\approx 0.3\,$s.

Earth is experiencing as a consequence of 3rd Newton's law the same force yet in opposite direction, thus 

$$F_{\mathrm{Erde}}=-5\,\mathrm{N}=m_{\mathrm{Erde}}\,a_{\mathrm{Erde}}$$

The accleration of Earth caused by the apple is $a_\mathrm{Erde}=8.3\cdot 10^{-27}\,$m/s$^2$. As a result during the the fall time of $0.3\,$s Earth moves by $3.8\cdot 10^{-28}\,$m towards the apple.

## 3.3 Kind of forces

### 3.3.1 Elastic forces, spring force

The linear spring generates a restoring force that is linear to the deflection $x$. The direction of the force is opposite to the deflection, which explains the negative sign in equation (4.3.1).


<img src="figs/4.3_1.png" width=270px>

$$\vec{F}_F=-k\,x\,\vec{e}_x\tag{4.3.1}$$

An elastic force is independent of the path taken for the deflection (only on the total deflection). This type of force is called a conservative force.

### 3.3.2 Friction

Is the force resisting the relative motion of solid surfaces sliding against each other:

* Static friction ($\vec{v}=0$), d.h. the body remains at rest)

<img src="figs/4.3_2.png" width=270px>

The static friction resists the applied force up to a maximum value. As long as this maximum value is not reached, the body remains at rest.

$$\vec{F}_{R}+\vec{F}=0$$

A normal force $\vec{F}_N=m\,g\,\vec{e}_y$ acts on the body, and only this determines the maximum static friction through a static friction coefficient $\mu_0$.

$$F_{R,max}=\mu_0\,F_N$$

The maximum static friction and thus the coefficient of friction can be easily determined on an inclined plane

<img src="figs/4.3_3.png" width=270px>

The force that moves the body downwards is 
$$F=m\,g\sin\alpha$$

and the normal force

$$F_N=m\,g\,\cos\alpha\quad .$$

The maximum static friction force is $F_{R,max}=-\mu_0\,F_N$. This means that once the force on the body is greater than $\left|F_{R,max}\right|$, then the body slides. Sliding therefore starts when

$$\left|F\right|=\left|F_{R,max}\right|$$

$$m\,g\,\sin\alpha=\mu_0\,F_N=\mu_0\,m\,g\cos\alpha$$

$$\tan\alpha=\mu_0$$

* Sliding friction

Once the body moves at a velocity $\vec{v}=v\,\vec{e}_v$, the dynamic frictional force acts on the body. This force points opposite to the direction of the velocity $\vec{e}_v$.  

<img src="figs/4.3_4.png" width=270px>

$$\vec{F}_{GL}=-\mu_{GL}\left|\vec{F}_N\right|\vec{e}_v$$

The unit vector can also be expressed as follows

$$\vec{e}_v=\frac{\vec{v}}{\left|\vec{v}\right|}$$

and thus

$$\vec{F}_{GL}=-\mu_{GL}\left|\vec{F}_N\right|\frac{\vec{v}}{\left|\vec{v}\right|}\quad .$$

$F_{GL}$ is independent of the magnitude of $\vec{v}$.

#### 3.4.1 Equivalence of mass

Let us apply Newton's 2nd law on an object where only gravity acts, i.e. no normal force affects the motion. Then we obtain

$$\sum F = m\,a$$
$$m\,g=m\,a$$
$$a=g$$

We find the acceleration of the object is independent on its mass.

##### 3.4.2 Accelerated frames of reference

The weight of a body is the force that the body exerts on the scale when it comes into contact with it. So a weight of 981 N is the force that a body with a mass of 100kg exerts on a scale. In principle, the scale consists of a spring that exerts such a strong force that the weight comes to rest. In common parlance and on scales, the weight is not expressed in Newtons, but in kilograms. As you will see below, this convention only really makes sense on earth and in an unaccelerated reference system.

<img src="figs/4.2_1.png" width=550px>

In the illustration above, the spring force of the scales indicates the weight of the body. This display is the weight of the body. How does the weight change in accelerated frames of reference?

*Left:*

The balance is in an inertial frame of reference, thus simply $F_W=m\,g$. 

*Center:*
The scale is in an elevator that is accelerated upwards (in the opposite direction to the acceleration due to gravity). Viewed from the outside (in the inertial system), we use Newton's 2nd law and obtain that the scale displays a force of $F_W=m(g+a)$, i.e. an increased weight. We feel heavier in an elevator, accelerating in the opposite direction to the earth.

*Right:*
The elevator is accelerated downwards (in the direction of the acceleration due to gravity). The weight is then $F_W=m(g-a)$. The body becomes lighter. If the elevator is accelerated downwards with $g$, i.e. the elevator is in free fall, then the scale no longer shows a weight. The body is weightless.  

##### 3.4.3 Atwood machine

<img src="figs/4.2_2.png" width=250px>

Two weights are attached to a massless rope. The rope lies on a massless and frictionless pulley. We want to determine the acceleration of the masses $m_1$ and $m_2$. To do this, we draw free-body diagrams of the two masses and determine the forces and accelerations acting on them.

*Left part of the rope*

$$T-m_1\,g=m_1\,a$$

$$T=m_1(a+g)$$

*Right part of the rope*

$$m_2\,g-T=m_2\,a$$

$$T=m_2(g-a)$$

After a short calculations we obtain the following results:

$$a=\frac{m_2-m_1}{m_1+m_2}\,g$$

$$T=\frac{2 m_1 \, m_2}{m_1+m_2}\,g$$
