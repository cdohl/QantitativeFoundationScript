### 8.3 Forced oscillations

#### Periodic excitation of oscillating systems

<img src="figs/24.3_1.png" alt="24.3_1" style="width: 200px;"/>

The suspension moves with  

$$\vec{u}(t)=\hat{u}\,\cos(\omega_E\cdot t)]\, \vec{e}_x$$

How does the mass $m$ move?

$$\vec{x}(t)=x(t)\,\vec{e}_x$$

The deflection of the spring is $x(t)-u(t)$, and a Stoke friction force acts

$$\underbrace{-k\left[x(t)-u(t)\right]}_{\mathrm{spring}}\,\vec{e}_x - \underbrace{6 \pi \eta R \dot{x}}_{\mathrm{friction}}\,\vec{e}_x=m\,\ddot{x}\,\vec{e}_x$$

⇨ rearrangement

$$\ddot{x}+\underbrace{\frac{6\pi\eta_R}{m}}_{2\,\delta}\,\dot{x}+\underbrace{\frac{k}{m}}_{\omega_0^2}\,x = \underbrace{\frac{k}{m}}_{\omega_0^2} \, u(t)$$

<div class=“alert alert-info”>
$$\ddot{x}+2\,\delta\,\dot{x}+\omega_0^2\,x = \omega_0^2 \, \hat{u}\,\cos(\omega_E\,t)\tag{8.3.1}$$
<p>
<center>inhomogeneous 2nd order ODE</center>
</div>


#### Solution of the ODE:

  $$x(t)=x_\mathrm{homogeneous}(t)+x_\mathrm{particular}(t)$$

  $x_\mathrm{homogeneous}(t)$ is already known:
  
  $$x_\mathrm{homogeneous}(t) = A \cdot e^{-\delta t} \, \cos\left(\omega t -\varphi_0\right)$$
  
  for $t\rightarrow \infty$ this solution will die out.
  
  ⇨ transient response
  
  ⇨ afterwards
  
  $$x(t) = x_\mathrm{particular}(t)$$
  
  $$x(t) = x_p(t)$$

#### Search for the special solution

  ⇨ Approach
  
  $$x_p(t)=\hat{x}_p \, \cos \left(\omega_E\,t - \varphi\right) $$
  
  ⇨ System moves with the circular frequency $\omega_E$
  
  $$\dot{x}_p= -\hat{x}_p \, \omega_E \, \sin\left(\omega_E\,t - \varphi\right)$$

  $$\ddot{x}_p= -\hat{x}_p \, \omega_E^2 \, \cos\left(\omega_E\,t - \varphi\right)$$
  
  ⇨ Insert into DGl and determine $\hat{x}_p$ and $\varphi$
  
  After a lengthy calculation ...
  
<p>
    
* Phase shift: $\varphi$

  $$\tan \varphi = \frac{2\,\delta\,\omega_E}{\omega_0^2-\omega_E^2}\tag{8.3.2}$$
  

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt


omega0 = 2.*np.pi #frequency 1 Hz
delta = [0.01,1.,2.] #damping coefficient .2 1/s
omegaE = np.linspace(0., 2*omega0 ,500)


fig = plt.figure(figsize=(8,5))
for d in delta:
    phi=np.arctan(2*d*omegaE/(omega0**2.-omegaE**2))%np.pi
    plt.plot(omegaE/omega0,phi,'k')
plt.gca().axvline(x=1., color='k', linestyle='--' )
plt.gca().axhline(y=np.pi, color='k', linestyle='--' )
plt.xlabel(r'$\omega_E$',fontsize=15)
plt.ylabel(r'$\phi$',fontsize=15)
plt.yticks([0,np.pi/2,np.pi], (r'0', r'$\pi/2$', r'$\pi$'))
plt.xticks([0,1.,2.], (r'0', r'$\omega_0$', r'$2\,\omega_0$'))
plt.text(1.1,1.0,r'$\delta\approx0$',fontsize=15)
plt.text(0.85,.5,r'$\delta_1$',fontsize=15)
plt.text(0.5,1.0,r'$\delta_2$',fontsize=15)
plt.text(0.3,2.5,r'$\delta_2>\delta_1$',fontsize=15)

plt.gca().tick_params(axis='both', which='major', labelsize=15)
plt.show()
```

* Amplitude response: $\hat{x}_p=\hat{x}_p(\omega_E)$

  $$\hat{x}_p=\frac{\omega_0^2\,\hat{u}}{\sqrt{\left(\omega_0^2-\omega_E^2\right)^2
  +\left(2\,\delta\,\omega_E\right)^2}}$$

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt


omega0 = 2.*np.pi #frequency 1 Hz
delta = [0.1,1.,2.] #damping coefficient .2 1/s
omegaE = np.linspace(0., 3*omega0 ,500)
u=1.

fig = plt.figure(figsize=(8,5))
for d in delta:
    xp=omega0**2.*u/np.sqrt((omega0**2.-omegaE**2.)**2.+(2*d*omegaE)**2.)
    plt.plot(omegaE/omega0,xp,'k')

    plt.ylim(0,5.*u)
plt.gca().axvline(x=1., color='k', linestyle='--' )

plt.xlabel(r'$\omega_E$',fontsize=15)
plt.ylabel(r'$\hat{x}_p$',fontsize=15)
plt.xticks([0,1.,2.,3.], (r'0', r'$\omega_0$', r'$2\,\omega_0$',r'$3\,\omega_0$'))
plt.yticks([1.,5.], (r'$\hat{u}$',''))
plt.text(1.3,4.,r'$\delta=0$',fontsize=15)
plt.text(0.85,3.3,r'$\delta_1$',fontsize=15)
plt.text(0.8,1.9,r'$\delta_2$',fontsize=15)
plt.text(0.15,4.,r'$\delta_2>\delta_1$',fontsize=15)

plt.gca().tick_params(axis='both', which='major', labelsize=15)
plt.show()
```

  $$\omega_E \rightarrow 0 \quad \mathrm{then}\quad \hat{x}_p=\hat{u}$$
  
  $$\omega_E \rightarrow \infty \quad \mathrm{then}\quad \hat{x}_p=0$$
  
* resonance frequency: $\omega_{Res}$

  At which excitation frequency $\omega_E$ is the response $\hat{x}_p(\omega_E)$ largest?
  
  $$\left.\frac{\mathrm{d}\,\hat{x}_p}{\mathrm{d}\,\omega_E}\right|_{\omega_E=\omega_{Res}}=0$$
  
  after a brief calculation ⇨ 
  
  $$\omega_{Res}=\sqrt{\omega_0^2-2\,\delta^2}$$
  
  ⇨ thus only if $\omega_0>2\,\delta^2$.
