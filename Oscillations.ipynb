## 8.1. Free, undamped harmonic oscillations

#### 8.1.1 Example: Spring oscillator

<img src="figs/24.1_3.png" alt="24.1_3" style="width: 400px;"/>

Spring force law:

$$\vec{F}_F=-k \cdot x\cdot \vec{e}_x$$

$$ \vec{F} = m\,\frac{\mathrm{d}^2 x}{\mathrm{d}\,t^2}\cdot \vec{e}_x=m\, \vec{a}$$

$$m\, \ddot{x} = - k \, x $$

or

$$\ddot{x}+\frac{k}{m}\, x = 0\tag{8.1.1}$$

This is a homogeneous 2nd order differential equation with constant coefficients.

#### 8.1.2 Example: Pendulum

A mathematical pendulum is a pendulum with length $l$ where a point mass $m$ is attached to a massless string. 

  <img src="figs/8n_1.png" alt="8n_1" style="width: 350px;"/>

The force tangent to the trajectory (circular section) is through geometric considerations

$$
F_\text{tan}=-m g \sin \Theta
$$

The negative sign comes from the fact that the force is opposite the positive angle $\Theta$.  This force leads to an acceleration of the point mass on the circular section. This acceleration is 

$$a=\frac{\text{d}}{\text{d}t}\left(l \omega\right)=l\dot{\omega}=l\ddot{\theta}
$$

Thus we obtain from force balance

$$
l\ddot{\theta}+g \sin{\Theta}=0
$$
or
$$
\ddot{\theta}+\frac{g}{l} \sin{\Theta}=0
$$

For the small angle approximation $\Theta \ll 1$ and thus $\sin\Theta \approx \Theta$ (check with Taylor series around $\Theta =0$) we can simplify the non-linear ODE to a linear ODE

$$
\ddot{\theta}+\frac{g}{l} \Theta=0\tag{8.1.2}\quad
$$

which resembles Eq. (8.1.1).


#### 8.1.2 Solution of the differential equation: Spring oscillator
  
Which function $x=x(t)$ fulfills the differential Eqs (8.1.1)?


Approach: $x(t)=A\, e^{\,\lambda\,t}$
  
Now we have to determine the parameters $A$ and $\lambda$:
  
\begin{eqnarray}
x(t)&=&A\cdot e ^{\lambda t}\\
\dot{x}(t)&=&A\cdot \lambda\cdot e ^{\lambda t}\\\
\ddot{x}(t)&=&A\cdot \lambda^2\cdot e ^{\lambda t}
\end{eqnarray}

  ⇨ Insert into the ODE (8.1.1)
  

\begin{eqnarray}
A\cdot \lambda^2\cdot e ^{\lambda t} + \frac{k}{m} A\cdot e ^{\lambda t} &=&0\\\
A\cdot e ^{\lambda t}\cdot\left[\lambda^2 +\frac{k}{m}\right] &=&0
\end{eqnarray}
can only be fulfilled if $\left[\lambda^2 +\frac{k}{m}\right]=0$. This is called the characteristic equation and we determine its solutions.

$$\lambda^2=-\frac{k}{m}$$

We now denote $\frac{k}{m}=\omega_0^2$, the reason will soon become clear.

$$\lambda_{1,2}=\pm \sqrt{-\omega_0^2}=\pm\sqrt{-1}\,\omega_0=\pm i \omega_0$$

$$
\left.
\begin{array}{11}
\lambda_1=+ i \omega_0\\\
\lambda_2=- i \omega_0
\end{array}
\right\} \mathrm{two\,solutions}
$$


And insert into the ansatz

$$x(t)=A_1 e^{i\omega_0 t} +A_2 e^{-i\omega_0 t}\tag{8.1.3}$$

Determination of $A_1$ and $A_2$ from the initial conditions (IC) 

For the spring oscillation, $x(t=0)=x_0$ and $\dot{x}(t=0)=0$. The last IC states that the initial velocity is zero.


IC 1: $x(t=0)=x_0=A_1 + A_2$

IC 2: $\dot{x}=A_1\cdot i \omega_0\, e^{i\omega_0 t} - A_2\cdot i \omega_0\, e^{-i\omega_0 t}$
  
  for $t=0$: $\dot{x}=0$ ⇨ $A_1 i \omega_0 - A_2 i \omega_0=0$
   
   ⇨ $A_1=A_2$
   
   ⇨ With IC 1: $A_1=\frac{x_0}{2}$ und $A_2=\frac{x_0}{2}$
   
   
Now inserting these constants into the general solution Eq. (8.1.3) we obtain
   $$x(t)=\frac{x_0}{2}\underbrace{\left( e^{i\omega_0 t}+e^{-i\omega_0 t}\right)}_{2\cos(\omega_0 t)}$$

$$x(t)=\frac{x_0}{2}\cdot 2\cdot \cos( \omega_0 t)$$

<div class="alert alert-success">
$$x(t)=x_0\cdot \cos( \omega_0 t)$$
</div>

for $x(t=0)=x_0$ und $\dot{x}(t=0)=0$.


* Every other IC leads to different $A_1$ and $A_2$ ⇨ phase shifts.

In general:

<div class="alert alert-success">
$$x(t)=x_0 \, \cos\left(\omega_0 \cdot t - \varphi_0\right)$$
</div>

  $\varphi_0$ is the phase shift.

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi,500)

def update(x0=1., phi0 = 0.,omega0 = 1.):
    T0=2.*np.pi/omega0
    fig = plt.figure(figsize=(7,4))
    plt.plot(t/T0,x0* np.cos( omega0*t - phi0))
    plt.text(0,.75,r'$x(t)=x_0 \cos(\omega_0 \cdot t - \varphi_0)$',fontsize=15)
    plt.ylim(-1.,1.)
    plt.xlabel(r'$t/T_0$',fontsize=15)
    plt.ylabel('amplitude',fontsize=15)
    plt.show()

interact(update,x0=FloatSlider(description=r'\(x_0\)', min=0., max=1.,value=0.5),\
                phi0=FloatSlider(description=r'\(\varphi_0\)', min=-3., max=3.,value=0.),\
                omega0=FloatSlider(description=r'\(\omega_0\)', min=.5, max=3.,value=1.));
```

#### 8.1.3 Interpretation

<div class="alert alert-success">
$$x(t_1)=x(t_1+T_0)$$
</div>

  $\omega_0=\frac{2\pi}{T_0}$
  
  $\omega_0=2\pi\cdot f_0$

  $f_0 = \frac{1}{T_0}$

  $\omega_0$: angular frequency
  
  $f_0$: frequency

  $T_0$: period duration


<div class="alert alert-info">General differential equation of a free undamped oscillation
    
$$ \ddot{x}+\frac{k}{m}\,x = 0$$

can be written as

$$ \ddot{x}+\omega_0^2\,x = 0$$

with 
$$\omega_0^2=\frac{k}{m}$$.
</div>



#### 11.1.5 Energy balance for the spring oscillator

  \begin{eqnarray}
  W_{ges}&=&W_{kin}+W_{pot}\\
  &=& \underbrace{\frac{m}{2}\dot{x}^2}_{\text{kinetic and}} + \underbrace{\frac{k}{2}x^2}_{\text{potential energy}}
  \end{eqnarray}
  
  $$ \dot{x}=-x_0\,\omega_0\,\sin(\omega_0\,t)=-x_0\cdot\sqrt{\frac{k}{m}}\,\sin(\omega_0\,t)$$
  
  \begin{eqnarray}
  W_{ges}&=&\frac{k}{2}\,x_0^2\,\sin^2(\omega_0\,t)+\frac{k}{2}\,x_0^2\,\cos^2(\omega_0\,t)\\
  &=& \frac{k}{2}\,x_0^2\underbrace{\left[\sin^2(\omega_0\,t)+\cos^2(\omega_0\,t)\right]}_1
  \end{eqnarray}
  
  <div class="alert alert-info">The total energy is constant, thus it is not a function of time!
  $$W_{ges}=\frac{k}{2}\,x_0^2=\text{const.}$$
  
  </div>

## 8.2 Damped harmonic oscillations

#### 8.2.1 Example: spring oscillation with friction

  Ball in a liquid with Stoke's friction
  
  $$\vec{F}_{fric.}=-6\,\pi\,\eta\,R\,\dot{x}\,\vec{e}_x$$
  
  Sum of all forces results in acceleration:
  
  $$\sum_{i=1}^n \vec{F}_i=\vec{F}_{spring}+\vec{F}_{fric.}=m\,\ddot{x}\,\vec{e}_x$$
  
  $$-k\,x\,\vec{e}_x-6\,\pi\,\eta\,R\,\dot{x}\,\vec{e}_x=m\,\ddot{x}\,\vec{e}_x$$
  
  multiply with $\vec{e}_x$ and divide by $m$:
  
  $$\ddot{x}+\frac{6\,\pi\,\eta\,R}{m}\,\dot{x}+\frac{k}{m}\,x = 0$$
  
  with $2\delta=\frac{6 \pi \eta R}{m}$ and $\frac{k}{m}=\omega_0^2$ we obtaine a a general harmonic damped ODE:
  
  <div class="alert alert-info">

  $$\ddot{x} + 2\,\delta\,\dot{x} + \omega_0^2\, x = 0$$ 
  
  </div>
  
#### 8.2.2 Solving the ODE:
  
  ⇨ Approach
  
  \begin{eqnarray}
  x(t)&=&A\cdot e^{\lambda\,t}\\\
  \dot{x}(t)&=&A\cdot\lambda\, e^{\lambda\,t}\\
  \ddot{x}(t)&=&A\cdot\lambda^2\, e^{\lambda\,t}
  \end{eqnarray}
  
  ⇨
Insert into DGl and factorize
  
  $$A\cdot e^{\lambda\,t}\cdot\underbrace{\left[\lambda^2+2\delta\lambda+\omega_0^2\right]}_0=0$$
  
The approach is only fulfilled if the expression in the brackets is zero:

 $$\lambda^2+2\delta\lambda+\omega_0^2=0$$
  ⇨
  \begin{eqnarray}
  \lambda_1=-\delta+\sqrt{\delta^2-\omega_0^2}\\
  \lambda_2=-\delta-\sqrt{\delta^2-\omega_0^2}\\
  \end{eqnarray}
  
  <div class="alert alert-info">
  <b>Full solution:
  $$
  x(t)= A_1\,e^{-\delta\,t}\,e^{+\sqrt{\delta^2-\omega_0^2}\,\cdot t}
      + A_2\,e^{-\delta\,t}\,e^{-\sqrt{\delta^2-\omega_0^2}\,\cdot t}\tag{8.2.1}$$
   </div>
   
   
* Distinguishing three cases depending on $\delta$ and $\omega_0$

  * if $\delta<\omega_0\quad \rightarrow$ oscillation case (weak damping)
  * if $\delta=\omega_0\quad \rightarrow$ aperiodic limit case
  * if $\delta>\omega_0\quad \rightarrow$ creep case (strong damping)

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi,500)


def update(x0=1., omega0 = 1., delta = .5):
    T0=2.*np.pi/omega0
    fig = plt.figure(figsize=(7,4))
    x=x0/2.*np.exp(-delta*t)*np.exp((delta**2-omega0**2+0j)**.5*t)+\
      x0/2.*np.exp(-delta*t)*np.exp(-(delta**2-omega0**2+0j)**.5*t)
    #in obiger Zeile braucht es +0j damit die Potenz von einer komplexen Zahl berechnet wird
    plt.plot(t/T0,np.real(x))
    plt.ylim(-1.,1.)
    plt.xlabel(r'$t/T_0$',fontsize=15)
    plt.ylabel('Amplitude',fontsize=15)
    plt.show()


interact(update,x0=FloatSlider(description=r'\(x_0\)', min=0., max=1.,value=0.5),\
                delta=FloatSlider(description=r'\(\delta\)', min=.0, max=3.,value=0.),\
                omega0=FloatSlider(description=r'\(\omega_0\)', min=.5, max=3.,value=1.));

```

#### 8.2.3 weak attenuation $\delta<\omega_0$

\begin{eqnarray}
\sqrt{\delta^2-\omega_0^2}&=&\sqrt{-1}\sqrt{\omega_0^2 -\delta^2}\\
&=& i \cdot \omega
\end{eqnarray}

Circular frequency of the damped oscillation:

$$\omega=\sqrt{\omega_0^2-\delta^2}$$
Allgemeine Lösung:
$$x(t)= A_1 \cdot e^{-\delta t} e^{i\omega t}+ A_2 \cdot e^{-\delta t} e^{-i\omega t}\tag{8.2.1a}$$


* Determination of $A_1$ and $A_2$ through the initial conditions (IC)

    * IC 1: $x(t=0)=x_0 \rightarrow x_0=A_1+A_2$
    * IC 2: $\dot{x}(t=0)=0$

<div class="alert alert-info">

Inserting into the ODE we obtain after a slightly lengthy calculation:

  $$x(t)=x_0 e^{-\delta t}\frac{\omega_0}{\omega}\cos \left(\omega t -\varphi_0\right)\tag{8.2.1b}$$
  
where the phase angle is given by
  
  $$\tan \varphi_0=\frac{\delta}{\omega}\quad .$$
  
</div>

* Comparison with the undamped case

  $$\delta = 0$$
  
  then is 
  
  $$\omega=\sqrt{\omega_0^2 - \delta^2}\quad \rightarrow \quad \omega =\omega_0$$ 
  
  and
  
  $$\tan \varphi_0 = \frac{\delta}{\omega_0}=\frac{0}{\omega_0} \quad \rightarrow \quad \varphi_0 = 0$$
  
  so that
  
  $$ x(t)=x_0 \cdot \cos(\omega_0 t )$$
  
  without damping is thus a special case of the damped oscillation case.
  
* Interpretation with damping:

  The time-dependent amplitude is
  
  $$x_0 \cdot \frac{\omega_0}{\omega}\cdot e^{-\delta t}$$
  
  * Logarithmic decrement $\Lambda$
  Let us take the solution of the damped oscillation Eq. (8.2.1b) 
  
  $$x(t)=x_0 e^{-\delta t}\frac{\omega_0}{\omega}\cos \left(\omega t -\varphi_0\right)$$
  
  and calculate the ratio of the amplitude of two consecutive oscillations we get
  
  $$\frac{x(t)}{x(t+T)}= e^{\delta T}\quad .$$
  
  The exponent $\delta \, T$ is called the logarithmic decrement $\Lambda$.
  
  $$\ln\left[\frac{x(t)}{x(t+T)}\right]=\delta \, T = \Lambda$$
  
  $\Lambda$ is the reciprocal of the oscillations at which the amplitude has fallen to $1/e\approx0.37$. 
  
* The degree of damping $D$ is defined as 

  $$D=\frac{\delta}{\omega_0}$$
  
* and the quality factor
  
  
  $$Q=\frac{\omega_0}{2\delta}=\frac{1}{2\,D}$$
  
  For attenuations that are not too large, the quality factor $Q$ can be conveniently expressed using the logarithmic decrement:
  
  $$Q=\frac{T \, \omega_0}{2 \Lambda}=\frac{T \, \pi}{T_0\, \Lambda}\approx\frac{\pi}{\Lambda}$$
  

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10 ,500)

x0 = 1. #amplitude 1
omega0 = 2.*np.pi #frequency 1 Hz
delta = .3 #damping coefficient .2 1/s

omega=np.sqrt(omega0**2.-delta**2.)
phi0=np.arctan(delta/omega)

x= x0*omega0/omega
T0=2.*np.pi/omega0
T=2.*np.pi/omega
x=x0*omega0/omega*np.exp(-delta*t/T0)*np.cos(omega*t/T0-phi0)

fig = plt.figure(figsize=(8,5))
plt.plot(t/T0,x,'k')
plt.plot(t/T0,np.exp(-delta *t/T0),'k-.',t/T0,-np.exp(-delta *t/T0),'k-.')
plt.gca().axhline(y=0, color='k')
plt.gca().axhline(y=1./np.exp(1.), color='g')
plt.gca().axvline(x=0, color='k')
plt.ylim(-1.,1.)
plt.xlabel(r'$t/T_0$',fontsize=15)
plt.ylabel('amplitude',fontsize=15)
plt.text(2,0.75,u'Envelope of of the oscillation $\propto e^{-\delta t}$',fontsize=15)
plt.show()
print("delta={0}".format(delta))
print("omega={0} und omega0={1}".format(omega,omega0))
print("T={0} und T0={1}".format(T,T0))
print("Lambda={0}, D={1}, Q={2}, Q approx {3}".format(T*delta,delta/omega0,omega0/2./delta,np.pi/T/delta))
```

#### 8.2.4 aperiodic limiting case $\delta=\omega_0$

Then
$$\omega=\sqrt{\omega_0^2-\delta^2}=0$$

To analyze the oscillation, a limit value analysis $\omega\rightarrow 0$ is necessary. To do this, we start with equation (8.2.1b)

$$ x(t)=x_0 e^{-\delta t}\left[\cos(\omega t)+\frac{\delta}{\omega}\sin (\omega t)\right]\quad .$$

The cosine term is simple:

$$\lim_{\omega\rightarrow 0} \cos (\omega t)=1$$

For the sine term, we can apply [de l'Hospital's rule](https://en.wikipedia.org/wiki/L%27H%C3%B4pital%27s_rule):

$$\lim_{\omega\rightarrow 0} \frac{\delta \sin (\omega t)}{\omega}=
\lim_{\omega\rightarrow 0} \frac{\delta t \cos (\omega t)}{1}=\delta \, t$$

and this results in the following for the aperiodic limit case

$$x(t)=x_0 \, e^{-\delta t}\left[ 1 + \delta \, t\right]$$

The system very quickly enters the rest state

* Applications:
  * Swing door
  * Shock absorber
  * Analog display instruments

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 16 * np.pi,500)

x0 = 1.

delta = .2
fig = plt.figure(figsize=(7,4))
x=x0*np.exp(-delta*t)*(1.+delta*t)

plt.plot(t*delta,x,'k')
plt.gca().axhline(y=0, color='k')
plt.gca().axvline(x=0, color='k')
plt.ylim(-.1,1.2)
plt.xlabel(r'$t\cdot\delta$',fontsize=15)
plt.ylabel('amplitude',fontsize=15)
plt.show()
print(r'x_0={0} and delta={1}'.format(x0, delta))
```

#### 8.2.5 Creep case $\delta>\omega_0$

We define $\sqrt{\delta^2 -\omega_0^2}=\delta'$

then Eq. (8.2.1) follows for the general solution:

$$x(t)= e^{-\delta\,t}\,\left( A_1\,e^{+\delta'\,\cdot t}+ A_2\,e^{-\delta'\,\cdot t}\right)$$

The exponential functions of $A_1$ and $A_2$ can be summarized. We obtain the time constants:
$$\frac{1}{\tau_1}=\delta-\delta'$$

$$\frac{1}{\tau_2}=\delta+\delta'$$

with 
$$\left[\tau_1\right]=\left[\tau_2\right]=1\mathsf{s}$$.


$$x(t)=A_1\,e^{-t/\tau_1}+A_2\,e^{-t/\tau_2}$$


* for strong attenuation:

  $$\delta \gg \omega_0$$ and $\dot{x}(t=0)=0$
  
  we can approximate $\delta'\approx\delta$, or $1/\tau_1\approx 0$ and $\tau_2\approx 2\delta$:
  
  $$x(t)\approx x_0 \, e^{-2\delta\,t}=x_0 \, e^{-t/\tau_2}$$

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 16 * np.pi,500)

x0 = 1.

tau_1=0.5
tau_2=0.05
fig = plt.figure(figsize=(7,4))
x=x0/2.*np.exp(-tau_1*t)+x0/2.*np.exp(-tau_2*t)

plt.plot(t,x,'k')
plt.gca().axhline(y=0, color='k')
plt.gca().axvline(x=0, color='k')
plt.ylim(-.1,1.2)
plt.xlabel(r'$t$',fontsize=15)
plt.ylabel('amplitude',fontsize=15)
plt.show()
print("x0={0}, tau1={1}, tau2={2}".format(x0, tau_1, tau_2))
```
